```python
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt, tensorflow as tf
from sklearn.model_selection import train_test_split
from PIL import Image

from tensorflow.keras import layers, Model
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

tf.random.set_seed(1337)
np.random.seed(1337)
```

```python
# =============================================================
# FASE 1  CARREGAR/CONSTRUIR SEU DATAFRAME (como voc j faz)
#           (exemplo simplificado baseado no seu cdigo)
# =============================================================
# Caminho absoluto ou relativo ao local onde seu Jupyter Notebook est rodando
base_path = os.path.expanduser('~/Downloads/archive (1)/flowers/flowers/Orange')  # <== AJUSTE

# Monta df_all com colunas: path, label
rows = []
for dirname, _, filenames in os.walk(base_path):
    jpgs = [f for f in filenames if f.lower().endswith('.jpg')]
    if not jpgs:
        continue
    folder_name = os.path.basename(dirname)
    for f in jpgs:
        rows.append({"path": os.path.join(dirname, f), "label": folder_name})

df_all = pd.DataFrame(rows)
print("Resumo por classe:\n", df_all['label'].value_counts().sort_index())

# BALANCEAMENTO (opcional)  ex.: igualar para 5175 por classe
controle_imagem_cat = 5175  # <== AJUSTE se quiser balancear
classes = sorted(df_all['label'].unique().tolist())

balanced = []
for c in classes:
    sub = df_all[df_all['label'] == c]
    n = min(controle_imagem_cat, len(sub))
    balanced.append(sub.sample(n, random_state=0))

df = pd.concat(balanced, ignore_index=True)
print("Aps balancear:\n", df['label'].value_counts().sort_index())

# SPLITS estratificados
n_classes = len(classes)
df_train, df_test = train_test_split(df, test_size=0.4, stratify=df['label'], random_state=0)
df_train, df_val  = train_test_split(df_train, test_size=0.5, stratify=df_train['label'], random_state=0)
print("Split sizes:", len(df_train), len(df_val), len(df_test))
```

```python
# =============================================================
# FASE 2  FUNO DE PLOTS (igual  sua, s agrupada)
# =============================================================

def plots(history, xlim=None, ylim=None):
  if hasattr(history, 'history'):
    history = history.history
  plt.figure(figsize=(14, 4))
  plt.subplot(1, 2, 1)
  plt.plot(history['loss'], '.-', label='Train loss')
  if 'val_loss' in history.keys():
    plt.plot(history['val_loss'], '.-', label='Val loss')
  plt.xlabel('Epochs'); plt.legend(); plt.yscale('log'); plt.grid(which='both')
  plt.subplot(1, 2, 2)
  plt.plot(history['accuracy'], '.-', label='Train accuracy')
  if 'val_accuracy' in history.keys():
    plt.plot(history['val_accuracy'], '.-', label='Val accuracy')
    best = np.max(history['val_accuracy']); last = history['val_accuracy'][-1]
    plt.title(f"Val acc: {best:.1%} (best) | {last:.1%} (last)")
  plt.xlabel('Epochs'); plt.legend(); plt.xlim(xlim); plt.ylim(ylim); plt.grid()
```

```python
# =============================================================
# FASE 3  CRIAR tf.data A PARTIR DO DATAFRAME
# Explicao: vamos ler caminho->bytes->JPEG->tensor, redimensionar e batch.
# O preprocess_input ser aplicado DENTRO do modelo (Lambda), para manter o pipeline limpo.
# =============================================================
#  AJUSTES
IMG_SIZE   = (224, 224)   # padro ResNet50
BATCH_SIZE = 64
n_classes  = len(classes)  # 'classes' deve existir (lista de nomes de classe)

# mapeia rtulo string -> ndice [0..n_classes-1]
class_to_idx = {c: i for i, c in enumerate(classes)}

def make_ds_from_df(df: pd.DataFrame, shuffle=False):
    paths  = df['path'].to_numpy()
    labels = df['label'].map(class_to_idx).to_numpy()

    ds = tf.data.Dataset.from_tensor_slices((paths, labels))

    def _load(path, y):
        img = tf.io.read_file(path)
        img = tf.io.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, IMG_SIZE)           # garante 224x224
        img = tf.cast(img, tf.float32)                 # preprocess_input cuidar da escala
        y   = tf.one_hot(y, depth=n_classes)           # <-- ONE-HOT AQUI
        return img, y

    ds = ds.map(_load, num_parallel_calls=tf.data.AUTOTUNE)
    if shuffle:
        ds = ds.shuffle(1024, seed=1337)
    ds = ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    return ds

train_ds = make_ds_from_df(df_train, shuffle=True)
val_ds   = make_ds_from_df(df_val)
test_ds  = make_ds_from_df(df_test)
```

```python

# =============================================================
# FASE 4  MODELO (AQUECIMENTO: BACKBONE CONGELADO)
# - Preprocessamento dentro do modelo (Lambda)
# - Usa 'categorical_crossentropy' (compatvel com ONE-HOT)
# =============================================================
from tqdm.keras import TqdmCallback
EPOCHS_WARM = 100

def build_model_frozen(num_classes: int) -> Model:
    # Como j redimensionamos no dataset, podemos fixar a entrada em 224x224
    inputs = Input(shape=IMG_SIZE + (3,), name='image')

    # (Opcional) Augmentao leve
    x = layers.RandomFlip('horizontal')(inputs)
    x = layers.RandomRotation(0.05)(x)
    x = layers.RandomZoom(0.1)(x)

    # Pr-processamento especfico da ResNet50 (NO muda rtulos)
    x = layers.Lambda(preprocess_input, name='preprocess')(x)

    # Backbone pr-treinado
    base = ResNet50(include_top=False, weights='imagenet', pooling='avg', name='resnet50')
    base.trainable = False
    x = base(x, training=False)  # BN em modo inferncia

    # Cabea leve
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = Model(inputs, outputs, name='ResNet50_frozen')
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss='categorical_crossentropy',   # <-- combina com ONE-HOT do dataset
        metrics=['accuracy']
    )
    return model

model = build_model_frozen(n_classes)
model.summary()

# Callbacks para o warmup
cb_warm = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, verbose=1),
    EarlyStopping(monitor='val_accuracy', patience=8, restore_best_weights=True),
    TqdmCallback(verbose=1)  # barras por batch/poca
]



# Treinamento (warmup)
history_warm = model.fit(
    train_ds,                   # usa os ds compatveis com ONE-HOT
    validation_data=val_ds,
    epochs=EPOCHS_WARM,
    callbacks=cb_warm,
    verbose=1
)

# Plotagem
plots(history_warm)
```

```python
# =============================================================
# FASE 5  FINE-TUNING PARCIAL (liberar ltimas N camadas)
# Explicao simples:
# - Agora destravamos s o topo da ResNet50 e treinamos com LR pequeno.
# - Isso refina as features para o seu domnio (laranjas Fresh/Rotten/Formalin-mixed).
# =============================================================

UNFREEZE_LAST_N = 50  # 30~100 costuma funcionar

# localizar o backbone dentro do modelo (camada ResNet50)
backbone = None
for layer in model.layers:
    if isinstance(layer, tf.keras.Model) and 'resnet50' in layer.name:
        backbone = layer
        break
if backbone is None:
    # fallback simples
    for layer in model.layers:
        if hasattr(layer, 'layers') and any('conv' in l.name for l in layer.layers):
            backbone = layer
            break

if backbone is not None:
    backbone.trainable = True
    for l in backbone.layers[:-UNFREEZE_LAST_N]:
        l.trainable = False
    for l in backbone.layers[-UNFREEZE_LAST_N:]:
        l.trainable = True

    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-5),  # LR pequeno  crucial no FT
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    cb_ft = [
        ModelCheckpoint('best_resnet50.keras', monitor='val_accuracy', save_best_only=True),
        EarlyStopping(patience=3, restore_best_weights=True, monitor='val_accuracy'),
        ReduceLROnPlateau(patience=2, factor=0.2, monitor='val_loss'),
    ]

    hist_ft = model.fit(train_ds, validation_data=val_ds, epochs=10, callbacks=cb_ft)
    plots(hist_ft)
else:
    print('[Aviso] No consegui localizar o backbone para liberar camadas  pulei o fine-tuning.')
```

```python
# =============================================================
# FASE 6  AVALIAO FINAL NO TESTE
# =============================================================
print('\nAvaliando no conjunto de teste...')
metrics = model.evaluate(test_ds, return_dict=True)
print({k: float(v) for k,v in metrics.items()})

# =============================================================
# FASE 7  PREDIO EM UMA IMAGEM (exemplo)
# =============================================================
# img_path = df_test['path'].iloc[0]
# img = tf.keras.utils.load_img(img_path, target_size=IMG_SIZE)
# x = tf.keras.utils.img_to_array(img)[None, ...]
# x = preprocess_input(x)
# probs = model.predict(x)
# pred = int(np.argmax(probs, axis=1)[0])
# print('Classe prevista:', classes[pred], '| Confiana:', float(np.max(probs)))

print('\nPronto! Agora o fluxo est ALINHADO ao seu cdigo com DataFrame + Colab,\ncom ResNet50 pr-treinada, aquecimento e fine-tuning parcial.')
```

```python
# ================================================
# MTRICAS COMPLETAS NO TESTE (Accuracy/Precision/Recall/F1)
# ================================================
# ===== Reconstruo segura de 'classes' =====
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_recall_fscore_support

def get_classes_safely():
    # melhor fonte: o dicionrio usado para criar os ndices
    if 'class_to_idx' in globals() and isinstance(class_to_idx, dict) and len(class_to_idx) > 0:
        # ordena por ndice (0..n-1)
        return [k for k, v in sorted(class_to_idx.items(), key=lambda kv: kv[1])]
    # alternativa: deduz do df_all (pode no casar com o treino se a ordem tiver mudado)
    if 'df_all' in globals():
        return sorted(df_all['label'].unique().tolist())
    raise RuntimeError(
        "No encontrei 'classes' nem 'class_to_idx' nem 'df_all'. "
        "Execute as clulas de criao do DataFrame e do mapeamento antes das mtricas."
    )

if 'classes' not in globals() or classes is None or len(classes) == 0:
    classes = get_classes_safely()

# ===== y_true / y_pred gerados a partir do test_ds =====
y_true = []
y_pred = []
for batch_x, batch_y in test_ds:
    true_idx = np.argmax(batch_y.numpy(), axis=1)
    probs = model.predict(batch_x, verbose=0)
    pred_idx = np.argmax(probs, axis=1)
    y_true.append(true_idx)
    y_pred.append(pred_idx)

y_true = np.concatenate(y_true)
y_pred = np.concatenate(y_pred)

# ===== mtricas agregadas =====
acc = accuracy_score(y_true, y_pred)
prec_w, rec_w, f1_w, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)
prec_m, rec_m, f1_m, _ = precision_recall_fscore_support(y_true, y_pred, average='macro', zero_division=0)

print(f"\nACCURACY (teste): {acc:.4f}")
print(f"MACRO     Precision: {prec_m:.4f} | Recall: {rec_m:.4f} | F1: {f1_m:.4f}")
print(f"WEIGHTED  Precision: {prec_w:.4f} | Recall: {rec_w:.4f} | F1: {f1_w:.4f}\n")

print(classification_report(y_true, y_pred, target_names=classes, zero_division=0))

# ===== Matriz de confuso =====
cm = confusion_matrix(y_true, y_pred)
cm_norm = confusion_matrix(y_true, y_pred, normalize='true')

# usamos os tamanhos da matriz para definir ticks com segurana
xticks = np.arange(cm.shape[1])
yticks = np.arange(cm.shape[0])

fig, ax = plt.subplots(figsize=(5,4))
im = ax.imshow(cm, cmap='Blues')
ax.set_title('Matriz de Confuso (contagens)')
ax.set_xticks(xticks); ax.set_yticks(yticks)
ax.set_xticklabels(classes, rotation=45, ha='right'); ax.set_yticklabels(classes)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='black', fontsize=9)
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(5,4))
im = ax.imshow(cm_norm, cmap='Blues', vmin=0, vmax=1)
ax.set_title('Matriz de Confuso (normalizada)')
ax.set_xticks(xticks); ax.set_yticks(yticks)
ax.set_xticklabels(classes, rotation=45, ha='right'); ax.set_yticklabels(classes)
for i in range(cm_norm.shape[0]):
    for j in range(cm_norm.shape[1]):
        ax.text(j, i, f"{cm_norm[i, j]:.2f}", ha='center', va='center', color='black', fontsize=9)
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()
```
